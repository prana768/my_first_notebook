### What is the difference between hardware and software?

**Hardware** refers to the physical components of a computer system that you can touch, such as the CPU, keyboard, or monitor. It's the tangible part that performs the computations. **Software**, on the other hand, is the set of instructions, programs, and data that tells the hardware what to do. It's intangible, and examples include operating systems like Windows or applications like a web browser.
